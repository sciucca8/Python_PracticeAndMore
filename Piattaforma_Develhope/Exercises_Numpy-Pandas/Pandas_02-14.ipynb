```python
import pandas as pd
import numpy as np
import matplotlib.cm as cm
import seaborn as sns

df = pd.read_csv("world_alcohol - world_alcohol.csv")
df
```

```python
df2 = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth ': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
    index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
df2
```

```python
# exe2 - Write a Pandas program to filter those records where WHO region matches with multiple values (Africa, Eastern Mediterranean, Europe) from world alcohol consumption dataset
regions = ["Africa", "Eastern Mediterranean", "Europe"]
result2 = df.loc[df["WHO region"].isin(regions)]
result2
```

```python
# exe3 - Write a Pandas program to filter those records which not appears in a given list from world alcohol consumption dataset.
ls_records = [3, 17, 56, 89]
result3 = df.iloc[[record for record in range(df.shape[0]) if record not in ls_records]]
result3
```

```python
# exe4 - Write a Pandas program to filter all records where the average consumption of beverages per person from .5 to 2.50 in world alcohol consumption dataset
mean_dv = df.groupby("WHO region")["Display Value"].mean() 
result4 = mean_dv.loc[(mean_dv >= 0.5) & (mean_dv <= 2.5)]
result4
```

```python
# exe5 - Write a Pandas program to find number of consumption of wine per person greater than 2 in world alcohol consumption dataset.
result5 = df.loc[(df["Beverage Types"] == "Wine") & (df["Display Value"] > 2)]
result5
```

```python
# exe6 - Write a Pandas program to filter rows based on row numbers ended with 0, like 0, 10, 20, 30 from world alcohol consumption dataset.
result6 = df.loc[[row for row in range(len(df)) if str(row)[-1] != "0"]]
result6
```

```python
# exe7 - Write a Pandas program to select consecutive columns and also select rows with Index label 0 to 9 with some columns from world alcohol consumption dataset.
rows = list(range(10))
result7 = df.loc[rows, "Year": "Country"]
result7
```

```python
# exe8 - Write a Pandas program to filter all records starting from the 2nd row, access every 5th row from world alcohol consumption dataset.
result8 = df.iloc[1::5]
result8
```

```python
# exe9 - Write a Pandas program to split the following dataframe into groups based on school code. Also check the type of GroupBy object.
result9 = df2.groupby("school_code")
print(type(result9))
result9
```

```python
# exe10 - Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.
result10 = df2.groupby("school_code")["age"].agg(["mean", "min", "max"])
result10
```

```python
# exe11 - Write a Pandas program to split the following given dataframe into groups based on school code and class.
result11 = df2.groupby(["school_code", "class"])
for school, group in result11:
    print(school, "\n", group)
```
